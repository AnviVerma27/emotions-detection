```python
import os
import cv2 as cv
import numpy as np
```

```python
face_cascade = cv.CascadeClassifier('C:/Users/hp/Desktop/openCV/Face recognition/haarcascade_frontalface_default.xml')

emotion = []
dir = r'C:/Users/hp/Desktop/openCV/Face recognition/New folder'
for i in os.listdir(dir):
    emotion.append(i)

features = []
labels = []
```

```python
def create_train():
    for feeling in emotion:
        path = os.path.join(dir,feeling)
        label = emotion.index(feeling)

        max_img = 100
        
        for idx, img in enumerate(os.listdir(path)):
            if idx>= max_img:
                break
            img_path = os.path.join(path,img)
            img_array = cv.imread(img_path)
            gray = cv.cvtColor(img_array,cv.COLOR_BGR2GRAY)
            image = cv.resize(gray,(80,80))
            features.append(image)
            labels.append(label)
```

```python
create_train()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(featuresets, labelset, test_size=0.3, random_state=20)
```

```python
import tensorflow as tf

model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(80,80)),
    tf.keras.layers.Dense(612,activation=tf.nn.relu),
    tf.keras.layers.Dense(10,activation=tf.nn.softmax)
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train,y_train, epochs=10)
model.evaluate(X_test,y_test)
```

```python
from keras.models import save_model

save_model(model,'handwritten digit recognition.h5')
```
